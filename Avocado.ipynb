##### Feature Engineering for Date Column

```python
df.Date = pd.to_datetime(df.Date)
```

```python
# extract month feature
months = df.Date.dt.month
```

```python
# extract day of month feature
day_of_months = df.Date.dt.day
```

```python
# extract year feature
year = df.Date.dt.year
```

```python
df['Day']=day_of_months
df['Month']=months
df['Year']=year
```

```python
df.head()
```

Dropping index column,Date and type as well

```python
df.drop(['Unnamed: 0','Date','type'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#As year is Float type, we will convert it to integer type
df.year=df.year.astype('int64')
```

```python
#Checking if year present in dataset and the one we extracted are equal
df['year'].equals(df['Year'])
```

```python
#Since, they are equal, we will drop one year column from our dataset
df.drop('year',axis=1,inplace=True)
```

```python
df.head()
```

```python
#Label Encoding the categorical variable
from sklearn.preprocessing import LabelEncoder
label_encoder=LabelEncoder()
df['region'] = label_encoder.fit_transform(df['region']) 
```

```python
#Removing outliers using z-score
from scipy.stats import zscore
z=np.abs(zscore(df))
```

```python
df=df[(z<3).all(axis=1)]
df.shape
```

After removing outliers, we have 1436 rows and 13 columns in our dataset

```python
#Separating x and y for model Evaluation
x=df.drop('AveragePrice',axis=1)
y=df[['AveragePrice']]
```

```python
#Checking for skewness
x.skew()
```

```python
for i in x.columns:
    if x[i].skew()>0.5:
        x[i]=np.cbrt(x[i])
    if x[i].skew()<-0.5:
         x[i]=np.square(x[i])

x.skew() 
```

```python
#lets bring all features to a common scale
from sklearn.preprocessing import MinMaxScaler
mm=MinMaxScaler()
x_=mm.fit_transform(x)
x_=pd.DataFrame(x_,columns=x.columns)
x_
```

```python
#Checking for multicollinearity using VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["variables"] = x_.columns
vif["VIF"] = [variance_inflation_factor(x_.values, i) for i in range(x_.shape[1])]

vif
```

Toatl Bags, Small Bags have high VIF

```python
x_.drop(['Total Bags'],axis=1,inplace=True)
```

### Finding the best Random State

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
maxAccu=0
maxRS=0
for i in range(1,200):
    x_train,x_test,y_train,y_test=train_test_split(x_,y,test_size=.20,random_state=i)
    lg=RandomForestRegressor()
    lg.fit(x_train,y_train)
    predrf=lg.predict(x_test)
    acc=r2_score(y_test,predrf)
    if acc>maxAccu:
        maxAccu=acc
        maxRS=i
print("Best accuracy is ",maxAccu," on Random_state ",maxRS)
```
